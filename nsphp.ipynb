Необходимые импорты:

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

Функция вычисления производных:

```python
def derivatives(y, t, alpha, beta):
    y1, y2 = y
    dy1_dt = 1 - y1*y2
    dy2_dt = alpha*y2*(y1 - (1 + beta)/(1 + beta*y2))
    return [dy1_dt, dy2_dt]
```

Начальные условия:

```python
alpha = 0.5
beta = 1
y0 = [1, 0.001]
```

Временные метки:

```python
t = np.linspace(0, 5, 50)
```

Численное решение системы уравнений и построение графиков:

```python
for alpha, beta in [(0.5, 1), (4, 0.2), (6, 0.2), (8, 0.5)]:
    sol = odeint(derivatives, y0, t, args=(alpha, beta))
    
    y1 = sol[:, 0]
    y2 = sol[:, 1]

    plt.subplot(1, 2, 1)
    plt.yscale('log')
    plt.plot(t, y1, color='orange')
    plt.xlabel('t')
    plt.ylabel('y1')
    plt.grid()
    
    plt.subplot(1, 2, 2)
    plt.yscale('log')
    plt.plot(t, y2, color='teal')
    plt.xlabel('t')
    plt.ylabel('y2')
    plt.grid()

    plt.title(f'Графики y1 и y2 в логарифмическом масштабе при alpha = {alpha}, beta = {beta}')
    plt.tight_layout()
    plt.show()
    
    plt.plot(y1, y2, color='darkgreen')
    plt.xlabel('y1')
    plt.ylabel('y2')
    plt.grid()
    plt.title('Фазовая диаграмма')
    plt.show()
```
